```python
import requests
import csv
import secrets

# Update this URL to point to your registration endpoint
base_url = "https://7df0-2409-4080-e4f-688b-5c9b-9ba6-3677-6a22.ngrok-free.app/register"  # <-- Replace with your actual URL

def generate_easy_password():
    """
    Generate a human-friendly password by combining a random word
    with a two-digit random number.
    """
    words = ["sunshine", "rainbow", "forest", "ocean", "mountain", "breeze", "daisy", "pearl", "amber", "copper"]
    word = secrets.choice(words)
    # Generate a random two-digit number between 10 and 99.
    number = str(secrets.randbelow(90) + 10)
    return word + number

def generate_easy_security_answer():
    """
    Generate a human-friendly security answer by combining a random word
    with a two-digit random number.
    """
    words = ["blue", "green", "red", "yellow", "purple", "orange", "silver", "gold"]
    word = secrets.choice(words)
    number = str(secrets.randbelow(90) + 10)
    return word + number

teams_data = []

# Create 30 teams: Team_1, Team_2, ..., Team_30
for i in range(1, 31):
    team_name = f"Team_{i}"
    password = generate_easy_password()
    leader_phone = "1234567890"
    security_question = "red"
    security_answer = generate_easy_security_answer()

    form_data = {
        "team_name": team_name,
        "password": password,
        "leader_phone": leader_phone,
        "security_question": security_question,
        "security_answer": security_answer
    }

    try:
        response = requests.post(base_url, data=form_data)
        if response.ok:
            print(f"Registered {team_name} successfully with password: {password} and security answer: {security_answer}")
        else:
            print(f"Failed to register {team_name}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Exception occurred while registering {team_name}: {e}")

    teams_data.append({
        "team_name": team_name,
        "password": password,
        "leader_phone": leader_phone,
        "security_question": security_question,
        "security_answer": security_answer
    })

# Save the registration details to a CSV file
csv_filename = "registered_teams.csv"
with open(csv_filename, mode="w", newline="") as csvfile:
    fieldnames = ["team_name", "password", "leader_phone", "security_question", "security_answer"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for team in teams_data:
        writer.writerow(team)

print(f"\nRegistration data for {len(teams_data)} teams has been saved to {csv_filename}.")
```
